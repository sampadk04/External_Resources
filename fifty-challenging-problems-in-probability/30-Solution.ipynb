# Evening the sales

- A bread salesman sells on average 20 cakes on a round of his route
- *What is the chance that he sells an even number of cakes?*
    - We assume the sales follow a Poisson distribution

_____

- Our intuition tells us it should be 0.5
    - Let's calculate some probabilities manually and plot them before we solve for the actual probability

```python
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
list_odd = []
list_even = []

for k in np.arange(40):
    P = poisson.pmf(k, 20)
    if k % 2 == 0:
        list_even.append(P)
        list_odd.append(0)
    else:
        list_even.append(0)
        list_odd.append(P)
```

```python
plt.plot(np.arange(40), np.cumsum(list_odd), label='odd')
plt.plot(np.arange(40), np.cumsum(list_even), label='even')
plt.legend()
plt.show()
```

- As we can see, they both seem to be converging to 0.5

- For the Poisson distribution, we know that:

$$
P(X=k) = \frac{\lambda^{k}e^{-\lambda}}{k!}
$$

- Since we expect to sell 20 cakes, we have:

$$
P(X=k) = \frac{20^{k}e^{-20}}{k!}
$$

- Since we want to calculate the probability that we sell an even number of cakes, we only consider $k$ values where $k=2j$ where $j$ is a positive integer

- Therefore:

$$
P(\text{Sell an even number of cakes}) = \sum_{j=0}^{\infty}\frac{20^{2j}e^{-20}}{2j!} = \frac{cosh(20)}{e^{20}} \approx 0.5
$$

- Where $cosh$ is the hyperbolic cosine
